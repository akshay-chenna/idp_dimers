## Cross RMSD for top 10 structures

```python
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np 
import matplotlib
from matplotlib import cm
from matplotlib.ticker import StrMethodFormatter
from matplotlib.colors import ListedColormap, LinearSegmentedColormap

path = './iRMS.txt'

rmsd_values = np.genfromtxt(path).astype(int)
rmsd_mat = np.zeros((rmsd_values[:,1].max(),rmsd_values[:,1].max()))
rmsd_mat[rmsd_values[:,0]-1,rmsd_values[:,1]-1] = rmsd_values[:,2]
rmsd_mat = rmsd_mat + rmsd_mat.T


custom_map = cm.get_cmap('jet')
newcolors = custom_map(np.linspace(0,1,6))
new_cmp = ListedColormap(newcolors)

matplotlib.rcParams.update({'font.size': 22})
plt.figure(figsize=(15,10)).patch.set_facecolor('white')
sb.heatmap(rmsd_mat, cmap=new_cmp, vmax=15)
plt.title("Interface RMSD")

plt.axhline(y=10, color='w', linestyle='-', linewidth=3)
plt.axhline(y=20, color='w', linestyle='-', linewidth=3)
plt.axhline(y=30, color='w', linestyle='-', linewidth=3)
plt.axhline(y=40, color='w', linestyle='-', linewidth=3)
plt.axvline(x=10, color='w', linestyle='-', linewidth=3)
plt.axvline(x=20, color='w', linestyle='-', linewidth=3)
plt.axvline(x=30, color='w', linestyle='-', linewidth=3)
plt.axvline(x=40, color='w', linestyle='-', linewidth=3)

plt.text(2,2,'it0-noair', color='w')
plt.text(12,4,'it0 \n weighted', color='w')
plt.text(22,6,'it1-noair \n all', color='w')
plt.text(31,6,'it1-noair_ \n top50 \n noairit0', color='w')
plt.text(42,8,'it1-noair_ \n top50 \n weighted \nit0', color='w')
#plt.text(8,26,'Cis', color='k')
#plt.text(26,26,'Trans', color='w')
```
