```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Plots of Rg

```python
rg = np.genfromtxt('rg.txt')[:,1]
rg = rg.reshape(-1, 5001)
```

```python
sns.heatmap(rg, cmap='jet')
plt.xlabel('# Frame (10 ps)')
plt.ylabel('# Trajectory')
plt.title('Rg (nm)')
```

```python
plt.figure(figsize=(5,3))
plt.plot(np.arange(rg.mean(axis=0).size),rg.mean(axis=0))
plt.fill_between(np.arange(rg.mean(axis=0).size),rg.mean(axis=0)-rg.std(axis=0), rg.mean(axis=0)+rg.std(axis=0), alpha=0.4)
plt.xlabel('Time (ns)')
plt.ylabel('Rg (nm)')
plt.xticks(np.linspace(0,5000,5),[str(i/100) for i in np.linspace(0,5000,5)]);
plt.savefig('rg.jpg',dpi=300, bbox_inches="tight")
```

# P1 and P2 vectors

```python
p1 = np.genfromtxt('p1.txt').reshape(-1,5001)
p2 = np.genfromtxt('p2.txt').reshape(-1,5001)
```
