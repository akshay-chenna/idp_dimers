# Calculate dot product

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cosine

for val in range(1,21):
    sasa = np.genfromtxt('data/deltargetsasa'+str(val)+'.txt')
    mmpbsa = np.genfromtxt('data/mmpbsa'+str(val)+'.txt', delimiter=" ")
    nsp7 = np.abs(np.genfromtxt('nsp7_cluster1area.txt'))
    design_mmpbsa = np.genfromtxt('topbe_mmpbsa.txt')*4.2

    mmpbsa = mmpbsa
    sasa = sasa.reshape((-1,70))

    RT = 310*8.314/1000
    partition_function = np.exp(-mmpbsa/RT).sum()
    weighted_sasa = sasa*np.exp(-mmpbsa/RT)[:,None]/partition_function
    sasa_vector = weighted_sasa.sum(axis=0)
    dot_product = 1 - cosine(sasa_vector,nsp7)
    dp_docked = np.empty([sasa.shape[0]])
    x = np.array([0.9,0.9])
    for i in range(0,sasa.shape[0]):
        dp_docked[i] = 1 - cosine(sasa[i,:],nsp7)
        #dp_docked[i,0] = i+1
    #np.savetxt('dp_2_poses.txt',dp_docked, fmt='%.0f \t %.3f')

    mmpbsa = np.concatenate((mmpbsa,np.array(design_mmpbsa[val-1:val])))
    dp_docked = np.concatenate((dp_docked, x[0:1]))
    plt.figure(figsize=(12,8))
    plt.rcParams['font.size'] = '16'
    plt.scatter(mmpbsa, dp_docked, alpha = 0.2, edgecolor='k')
    plt.xlim([mmpbsa.min()-20,100])
    plt.xlabel('MMPBSA (kJ/mol)')
    plt.ylabel('Dot product')
    plt.title('Rigid global docking, Design ' +str(val)+' dot product='+str(dot_product)[:4])
    plt.savefig('data/MMPBSA_dp'+str(val)+'.jpg', dpi=300)
```

# Determine binding quality (MMPBSA)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cosine
for val in range(1,21):
    sasa = np.genfromtxt('data/deltargetsasa'+str(val)+'.txt')
    mmpbsa = np.genfromtxt('data/mmpbsa'+str(val)+'.txt', delimiter=" ")
    nsp7 = np.abs(np.genfromtxt('nsp7_cluster1area.txt'))
    design_mmpbsa = np.genfromtxt('topbe_mmpbsa.txt')*4.2

    mmpbsa = mmpbsa
    sasa = sasa.reshape((-1,70))

    RT = 310*8.314/1000

    dp_docked = np.empty([sasa.shape[0]])
    #x = np.array([0.9,0.9])
    x = np.array([0.1,0.1])
    for i in range(0,sasa.shape[0]):
        #dp_docked[i] = 1 - cosine(sasa[i,:],nsp7)
        dp_docked[i] = cosine(sasa[i,:],nsp7)*10

    mmpbsa = np.concatenate((mmpbsa,np.array(design_mmpbsa[val-1:val])))
    dp_docked = np.concatenate((dp_docked, x[0:1]))
    
    pf = np.exp(-mmpbsa/RT).sum()
    cutoff = 0.3*10
    binding_quality = (np.exp(-(dp_docked[:,None]**2)/(cutoff)**2)*(np.exp(-mmpbsa/RT)[:,None])/pf).sum()
    print('The binding quality of Design %d is %.2f' %(val,binding_quality))
```

# Determing binding quality (REU)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cosine
for val in range(2,3):
    sasa = np.genfromtxt('../reu_docked/outputs'+str(val)+'/deltargetsasa'+str(val)+'.txt')
    mmpbsa = np.genfromtxt('../reu_docked/outputs'+str(val)+'/reu'+str(val)+'.txt', delimiter=" ")*4.2
    nsp7 = np.abs(np.genfromtxt('nsp7_cluster1area.txt'))
    design_mmpbsa = np.genfromtxt('../reu_docked/top_poses/reu_design.txt')*4.2

    mmpbsa = mmpbsa
    sasa = sasa.reshape((-1,70))

    RT = 310*8.314/1000

    dp_docked = np.empty([sasa.shape[0]])
    x = np.array([0.9,0.9])
    #x = np.array([0.1,0.1])
    for i in range(0,sasa.shape[0]):
        #dp_docked[i] = 1 - cosine(sasa[i,:],nsp7)
        dp_docked[i] = cosine(sasa[i,:],nsp7) * 10

    mmpbsa = np.concatenate((mmpbsa,np.array(design_mmpbsa[val-1:val])))
    dp_docked = np.concatenate((dp_docked, x[0:1]))
    
    nan = np.argwhere(np.isnan(dp_docked))
    dp_docked = np.delete(dp_docked,nan)
    mmpbsa = np.delete(mmpbsa,nan)
    
    pf = np.exp(-mmpbsa/RT).sum()
    cutoff = 0.9*10
    binding_quality = (np.exp(-(dp_docked[:,None]**2)/(cutoff)**2)*(np.exp(-mmpbsa/RT)[:,None])/pf).sum()
    print('The binding quality of Design %d is %.2f' %(val,binding_quality))
```
